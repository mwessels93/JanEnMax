To process wav-files we can use the program soundAnnotator.py. This is a script written using the libsoundannotator library.

You can run it by providing the script name as an argument to python:
    python pySoundAnnotator.py 

```bash
python soundAnnotator.py  -c --noofscales 100 --inputrate 44100 --outdir ~/data --decimation 2
```
